
# Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import the dataset

```python
ds=pd.read_csv('Position_Salaries.csv')
x=ds.iloc[:,1:-1].values
y=ds.iloc[:,-1].values
```

# Training the Random forest regression on whole dataset

Random forest is as same as decision tree 
and it take more number of trees to get an accurate prediction on the tree which is trained

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(x,y)
```


# Predicting the new result

```python
y_pred=regressor.predict([[6.5]])
y_pred
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,regressor.predict(x),color='blue')
plt.title("Decision tree regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

# Visualizing the decison tree regression results(higher resolution)

```python
x_grid=np.arange(min(x),max(x),0.1)
x_grid=x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,color='red')
plt.plot(x_grid,regressor.predict(x_grid),color='blue')
plt.title("Decision tree regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```
